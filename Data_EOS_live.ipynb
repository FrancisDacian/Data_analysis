```python
from bs4 import BeautifulSoup
import urllib2
import re
import random
import time
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains

quote_page = 'https://eosindex.io/projects'
element_id = "button.btn.btn-rounded.btn-noborder.btn-alt-primary"

developing_xpath = "//*[@id='filtersBlock']/div/div/div[3]/div/div[2]/div/div[2]/div[1]/label"
live_xpath = "//*[@id='filtersBlock']/div/div/div[3]/div/div[2]/div/div[1]/div[1]/label"
beta_xpath = "//*[@id='filtersBlock']/div/div/div[3]/div/div[2]/div/div[1]/div[2]/label"
concept_xpath = "//*[@id='filtersBlock']/div/div/div[3]/div/div[2]/div/div[2]/div[2]/label"

name_list_live = []

browser = webdriver.Firefox()
browser.maximize_window()
time.sleep(1)
browser.get(quote_page)

def smart_choose(element):
    for i in range(50):
        if is_element_exist(element):
            browser.find_element_by_css_selector(element).click()
            time.sleep(1)
            print("click successfully")
        else:
            print("no more buttons to click")
            break

def click_status(element):
        if is_status_exist(element):
            browser.find_element_by_xpath(element).click()
            time.sleep(1)
            print("click status successfully")
        else:
            print("no live checkbox to click")


def is_element_exist(element):
    try:
        browser.find_element_by_css_selector(element)
        return True
    except:
        return False

def is_status_exist(element):
    try:
        browser.find_element_by_xpath(element)
        return True
    except:
        return False
#choose the status you want to check  
#1.live_xpath
#2.developing_xpath
#3.beta_xpath
#4.concept_xpath
click_status(concept_xpath)

smart_choose(element_id)

PageSource = browser.page_source
browser.close()
soup = BeautifulSoup(PageSource)
name_box = soup.find_all("button",class_="btn-secondary")

for tag in name_box:
    tag_no_space = tag.string.strip()
    name_list_live.append(tag_no_space)
    print(tag_no_space)

```
