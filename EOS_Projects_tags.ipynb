```python
from bs4 import BeautifulSoup
from selenium import webdriver
import urllib2
import re
import random
import time
import pandas as pd
import xlwt

quote_page = 'https://eosindex.io/projects'
element_id = "button.btn.btn-rounded.btn-noborder.btn-alt-primary"

name_list = []
link_list = []
name_content = {}

browser = webdriver.Firefox()
browser.maximize_window()
time.sleep(1)
browser.get(quote_page)


def smart_choose(element):
    for i in range(50):
        if is_element_exist(element):
            browser.find_element_by_css_selector(element).click()
            time.sleep(1)
            print("click successfully")
        else:
            print("no more buttons to click")
            break
    
def is_element_exist(element):
    try:
        browser.find_element_by_css_selector(element)
        return True
    except:
        return False

smart_choose(element_id)

PageSource = browser.page_source
browser.close()
soup = BeautifulSoup(PageSource)

name_box = soup.find_all("button",class_="btn-secondary")
link_box = soup.find_all("a",class_="block block-link-shadow text-center ribbon ribbon-bookmark ribbon-crystal block-rounded block-bordered d-flex align-items-start flex-column full-height")

for tag in name_box:
    tag_no_space = tag.string.strip()
    name_list.append(tag_no_space)

for link in link_box:
    link_list.append(link['href'])
    print link['href']

for i in range(len(name_list)):
    name_content[i] = {'name':name_list[i],'link':"https://eosindex.io"+link_list[i]}

df_name_content = pd.DataFrame(name_content)
df_name_content_T = df_name_content.T

tags_content = {}

browser = webdriver.Firefox()
browser.maximize_window()

for web_index in range(len(name_list)):
    weblink = str(df_name_content_T.iat[web_index,0])
    browser.get(weblink)
    PageSource = browser.page_source
    soup = BeautifulSoup(PageSource)

    tag_box = soup.find_all("span",class_="badge badge-pill badge-primary mr-1")
    tag_content = []
    for tag_index in tag_box:
        
        tag_no_space = tag_index.string.strip()
        tag_content.append(tag_no_space)
        tags_content[web_index] = {"name":df_name_content_T.iat[web_index,1],"tags":tag_content}

browser.close()
df_tags_content = pd.DataFrame(tags_content)
df_tags_content_T = df_tags_content.T
#df_tags_content_T.to_excel('EOS_tags.xls')
```
